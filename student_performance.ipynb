# Student Mathematics Progress

```python
import pandas as pd
import numpy as np
```

```python
st_df = pd.read_csv(r'C:\Users\Abhin\OneDrive\Desktop\Machine Learning Projects\student performance\student\student-mat.csv', sep=';')
```

# Converting textual data into numerical data

```python
def binary(text):
    if text ==  "yes":
        return 1
    else:
        return 0
st_df['famsup'] = st_df['famsup'].apply(binary)
```

```python
st_df['Pstatus'] = st_df['Pstatus'].apply(binary)
```

```python
st_df['schoolsup'] = st_df['schoolsup'].apply(binary)
```

```python
st_df['paid'] = st_df['paid'].apply(binary)
```

```python
st_df['internet'] = st_df['internet'].apply(binary)
```

```python
st_df['romantic'] = st_df['romantic'].apply(binary)
```

# Selecting Important features from dataset and set the target column 

```python
X = st_df[[ 'studytime',  'absences','schoolsup', 'famsup', 'Medu', 'Fedu',  'romantic', 'goout', 'paid','famrel','health',  'Pstatus','G1','G2']]  
y = st_df['G3']
```

```python
X.head()
```

# Divide the data into training part and testing part

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

# Create an object of Linear Regression Model and fit the model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

# Predict the model and watch the difference between real values and predicted values

```python
y_pred = regressor.predict(X_test)
df = pd.DataFrame({'Real Values':y_test, 'Predicted Values':y_pred})
print(df)
```

# Predicted Model Score

```python
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error, r2_score
# Assuming you have your true y values (y_test) and predicted y values (y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Error (MAE):", mae)
print("R-squared (R2):", r2)
```
